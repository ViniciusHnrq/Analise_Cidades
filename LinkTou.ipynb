```python
# Importando a bibliotecas
import json
import pandas as pd
import requests
import re
```

```python
# Função para limpar os dados 

def limpar_dados(dados):
    # Procura por valores em branco 
    for chave, valor in list(dados.items()):
        if valor == "":
            del dados[chave]
    return dados 

    # Função para buscar os dados do CEP
    
def dados_cep(cep):

    # Aqui basta substituir o CEP da url para o CEP desejado
    url = f'https://viacep.com.br/ws/{cep}/json/' 
    
    # Fazendo a solicitação GET para obter os dados do CEP
    response = requests.get(url)

    # Verificando se a solicitação funcionou
    if response.status_code == 200:
        # Convertendo os dados da resposta para formato JSON e retirando os dados em branco com a função limpar_dados
        dados = response.json()
        dados_tratados = limpar_dados(dados)
        return dados_tratados
    
    else:
       return None       
```

```python
# São Roque
cep = 18143016
dados_do_cep = print(dados_cep(cep))
```

```python
# Juntando todos os DataFrame das cidades
df_todos = pd.concat([df_ibiuna, df_piedade, df_Sroque])

# Removendo dados indesejados na tabela 
df_todos['Valor'] = df_todos['Valor'].apply(lambda x: re.sub(r'\s\[\d{4}\]','', x))
df_todos['Valor'] = df_todos['Valor'].str.replace(r'hab/km²|km²|\(×1000\)|R\$|pessoas', '', regex=True)

# Transformando a tabela para um melhor desenvolvimento no bi, aqui deixo uma linha para cada cidade
df_todos_pivot = df_todos.pivot(index='Cidade', columns='Indicador', values='Valor')

# Resetando o índice para tornar a coluna 'Cidade' em uma coluna regular
df_todos_pivot.reset_index(inplace=True)

# Renomeando colunas 
df_todos_pivot.rename(columns={'Densidade demográfica ': 'Densidade demográfica (hab/km²)'}, inplace=True)
df_todos_pivot.rename(columns={'Área Territorial ': 'Área Territorial (km²)'}, inplace=True)
df_todos_pivot.rename(columns={'Receitas realizadas ': 'Receitas realizadas em R$ (x1000)'}, inplace=True)

# Teste
df_todos_pivot
# Criando arquivo em xlsx
#cidades = 'Dados_Tratados.xlsx'
#df_todos_pivot.to_excel(cidades, index=False)
```
